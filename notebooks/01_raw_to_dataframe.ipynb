# Converting Raw Scraped Data into a Pandas Dataframe
1. Read the csv file

```python
import pandas as pd

df = pd.read_csv("../data/raw/players.csv")
df.head()
df.info()
```

2. Remove unwanted columns

```python
cols_to_drop = [
    "Hit The Woodwork",
    "Hit The Woodwork (Per90)",
    "Hat-tricks",
    "3 or More Goals",
    "2 or More Goals",
    "Annual Salary (Per90)",
    "Red Cards (Per90)",
    "Yellow Cards (Per90)",
    "Matches Subbed Out",
    "Matches Subbed Out (Per90)",
    "Matches Subbed In",
    "Matches Subbed In (Per90)",
    "Matches Started (Per90)",
    "Matches Played (Per90)",
    
    "Minutes (Per90)",
    "Minutes Per Assist",
    "Minutes Per Goal",
    "Minutes Per Booking",
    "Minutes Per Goal Conceded",
    
    "Goals at Home (Per90)",
    "Goals at Away (Per90)",
    "Non-Penalty xG (npxG) (Per90)",
    "Goals Conceded (Per90)",
    "Clean Sheets (Per90)",
    "Save Percentage"
]

df = df.drop(columns=cols_to_drop)
df
```

3. Remove players without stats

```python
df = df[df["Footystats URL"].notna()]
df = df.drop(columns=["Footystats URL"])

df.rename(columns={"Save Percentage (Per90)" : "Save Percentage"}, inplace=True)

df.to_pickle("../data/raw/raw_df.pkl")
df
```
