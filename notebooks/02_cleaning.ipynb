Convert Money Columns

```python
money_cols = [
    "Market Value",
    "Annual Salary"
]

for col in money_cols:
    df[col] = df[col].apply(money_to_int)
```

Convert Percent Columns

```python
percent_cols = [
    "Pass Completion Rate",
    "Dribble Success Rate",
    "Cross Completion Rate",
    "Save Percentage",
    "Cards Over 0.5 (Per90)",
    "Penalty Conversion Rate",
    "Shot Accuracy",
]

for col in percent_cols:
    df[col] = df[col].apply(percent_to_float)
```

Convert "No" strings to numeric 0

```python
df["Penalty Conversion Rate"] = df["Penalty Conversion Rate"].replace("No Pens", np.nan)
df["Dribbles"] = df["Dribbles"].replace("Not Recorded", np.nan)
```

### 3. Data Type Conversion

Columns that should be Integers

```python
int_cols = [
    "Age",
    "Matches Played",
    "Minutes",
    "Matches Started",
    "Goals Scored",
    "Goal Involvement",
    "Goals at Home",
    "Goals at Away",
    "Goals Conceded",
    "Clean Sheets",
    "Tackles",
    "Interceptions",
    "Clearances",
    "Shots Blocked",
    "Dribbled Past",
    "Penalties Given Away",
    "Ground Duels",
    "Ground Duels Won",
    "Aerial Duels Won",
    "Assists",
    "Passes",
    "Successful Passes",
    "Key Passes",
    "Crosses",
    "Successful Crosses",
    "Dribbles",
    "Successful Dribbles",
    "Dispossesed",
    "Offsides",
    "Yellow Cards",
    "Red Cards",
    "Total Cards",
    "Cards Over 0.5",
    "Fouls Committed",
    "Fouled Against",
    "Penalties Scored",
    "Penalties Taken",
    "Penalties Missed",
    "Shots Faced",
    "Saves",
    "Punches",
    "Shots Taken",
    "Shots On Target",
    "Shots Off Target"
]
```

Test the columns for non-numeric data

```python
bad_values = {}

for col in int_cols:
    coerced = pd.to_numeric(df[col], errors="coerce")
    mask = coerced.isna() & df[col].notna()

    bad_values[col] = df.loc[mask, col].unique()

# bad_values
# df.loc[df["Dribbles"].eq("Not Recorded"), ["Name", "Position", "Dribbles"]]
```

Convert to Int64

```python
df[int_cols] = df[int_cols].astype("Int64")
```

### 3. Split according to Position

```python
df_gk = df[df["Position"] == "Goalkeeper"]
df_outfield = df[df["Position"] != "Goalkeeper"]
```

### 4. Remove Useless Columns

```python
useless_gk_cols = df_gk.columns[
    df_gk.apply(lambda col: ((col == 0) | (col.isna())).all())
]

useless_gk_cols
```

```python
cols_to_drop = [
    'Goals Scored',
    'Goals Scored (Per90)',
    'Goal Involvement',
    'Goal Involvement (Per90)',
    'Goals at Home',
    'Goals at Away',
    'Expected Goals (xG)',
    'Expected Goals (xG) (Per90)',
    'Non-Penalty xG (npxG)'
]

df_gk = df_gk.drop(columns=cols_to_drop, errors='ignore')
```

```python
useless_outfield_cols = [
    "Saves",
    "Saves (Per90)",
    "Save Percentage",
    "Punches",
    "Punches (Per90)",
    "Penalty Conversion Rate"
]

useless_outfield_cols
```

```python
df_gk = df_gk.drop(columns=useless_gk_cols)
df_outfield = df_outfield.drop(columns=useless_outfield_cols)
```

### 5. Drop NaN records

```python
df_outfield[df_outfield["Dribbles"].isna()]
```

```python
df_outfield = df_outfield.dropna(subset=["Dribbles"])
```
